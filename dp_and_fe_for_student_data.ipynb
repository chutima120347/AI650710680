```python
import pandas as pd
```

เป็นคำสั่งของ pandas ที่ใช้ในการ อ่านไฟล์ csv เข้ามาเป็น DataFrame เพื่อนําข้อมูลมาใช้ และ student.csv เป็นชื่อไฟล์ csv ที่จะถูกอ่าน

```python
df = pd.read_csv("student.csv")
```

```python
df
```

ลบแถวทั้งหมดที่มีค่า NaN ออกจาก DataFrame

```python
df_clean = df.dropna()
df_clean
df = df_clean.copy()
df
```

แก้ไขข้อมูลชื่อคณะให้เป็นมาตรฐานเดียวกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in['วิศวกรรมศาสตร์', 'วิศวฯ', 'วิศวกรรม', 'วิศวะ', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

แก้ไขข้อมูลเพศให้อยู่ในรูปแบบมาตรฐานเดียวกัน

```python
def standardize_gender(name):
    if name in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif name in['หญิง', 'ญ', 'F',  'Female']:
        return 'หญิง'
    else:
        return name
```

```python
df['เพศ']  = df['เพศ'].apply(standardize_gender)
```

```python
df
```

การจัดการค่าผิดปกติ โดยใช้ IQR เพื่อหาขอบเขตของข้อมูลปกติและกรองค่าผิดปกติ

```python
# วิธีง่ายๆ - ใช้ IQR (Interquartile Range)
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# กรองค่าผิดปกติ
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df
```

การจัดการค่าผิดปกติ โดยการแทนค่าอายุที่เกิน 100 ปี ด้วยค่ามัธยฐาน

```python
df.loc[df['อายุ'] > 100, 'อายุ'] = df['อายุ'].median()
df
```

กำหนดให้เกรดอยู่ในช่วง 0.00 ถึง 4.00

```python
df['GPA'] = df['GPA'].clip(lower=0, upper=4)
df
```

คำนวณค่า BMI จากน้ำหนักและส่วนสูง

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
```

```python
df
```

กำหนดจัดกลุ่มข้อมูล BMI เป็นหมวดหมู่สุขภาพตามช่วงค่า BMI

```python
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

```python
df
```
